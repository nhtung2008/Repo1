### Vẽ Tam giác lồng nhau </p>
cho 2 biến a và b là số nguyên 
Hãy vẽ a hình tam giác đều theo hình sau , với hình trong cùng có kích thước cạnh là b , các hình tiếp theo có kích thước tăng dần 20%


```python
import turtle as t
p=t.Turtle()
a=int(input('Nhập số hình tam giác: '))
b=int(input('Nhập kích thước cạnh tam giác: '))
for i in range(a):
    for j in range (3):
        p.fd(b)
        p.lt(120)
    p.penup()
    p.bk(b*0.1)
    b=b*1.2
    p.pendown()
t.done()
```

### Vẽ hình vuông lồng nhau
cho 2 biến a và b nguyên </p>
Hãy vẽ a hình vuông với kích thước lớn dần theo hình sau (mỗi hình vuông có các cạnh lớn hơn 10 đơn vị ) , hình vuông trong cùng có kích cỡ là b

```python
import turtle as t
p=t.Turtle()
a=int(input('Nhập số hình vuông: '))
b=int(input('Nhập kích thước cạnh hình vuông: '))
# toạ độ x,y của góc trái bên dưới của hình vuông đầu tiên:
x=0
y=0
for i in range(a):
    for j in range(4):
        p.fd(b)
        p.lt(90)
    p.penup()
    p.goto(x-5, y-5)
    b+=10
    x-=5
    y-=5
    p.pendown()
t.done()
```

### Vẽ hình
cho 2 biến nguyên a và b 
với b là kích thước cạnh của hình và a là số lượng hình


```python
import turtle as t
a=int(input('Nhập số hình bát giác đều cần vẽ: '))
b=int(input('Nhập độ dài cạnh hình bát giác đều: '))
p=t.Turtle()
for i in range(a):
    for j in range(8):
        p.fd(b)
        p.rt(45)
    p.rt(360/a)
t.done()
```

### Tính diện tích
Có một mảnh đất như hình dưới , cho biết các cạnh tương ứng và các hình như dưới . 
Các bạn hãy vận dụng kiến thức để tính chu vi,diện tích từng hình và diện tích tổng của mảnh đất.

```python
import math as m
a=10
b=30
c=50
d=100
cvhv=a*4
dthv=a*a
cvtg=a+b+c+m.sqrt((a+c)**2+b**2)
dttg=(a+c)*b/2
cvhcn=(c+d)*2
dthcn=c*d
print('Chu vi hình vuông: ',cvhv)
print('Diện tích hình vuông: ',dthv)
print('Chu vi tam giác: ', cvtg)
print('Diện tích tam giác: ',dttg)
print('Chu vi hình chữ nhật: ',cvhcn)
print('Diện tích hình chữ nhật: ',dthcn)
```

### hình tròn
